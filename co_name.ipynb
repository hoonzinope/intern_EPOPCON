```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
req = requests.get('http://blog.naver.com/PostView.nhn?blogId=9813028k&logNo=220814957558')
html = req.text
soup = BeautifulSoup(html, 'html.parser')
texts = soup.find("div",{"id": "post-view220814957558"}).get_text()
texts
```

```python
type(texts)
```

```python
texts = texts.split('\xa0')
len(texts)
```

```python
result = []
for t in texts:
    if '점' in t:
        result.append(t)
result
```

```python
name = []
for r in result:
    r = r.split(' ')
    for n in r:
        if '점' in n:
            name.append(n)
            
            
name
```

```python
name_r = []
for n in name:
    if '전화번호' in n:
        name_r.append(n.replace('전화번호',''))
    else:
        name_r.append(n)
name_r
```

```python
pd.DataFrame(name_r, columns = ['location_info']).to_pickle('location_name.pkl')
```
